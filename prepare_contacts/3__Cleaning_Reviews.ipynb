# Cleaning Reviews

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/grp_public_reporting.csv')
df = df.rename(columns=lambda x: x.strip())
df
```

```python
df = df.dropna(subset = ['star_value'], ignore_index=True)
df = df.drop_duplicates(subset = ['org_PAC_ID'], ignore_index=True)
df
```

```python
df.to_parquet('../data/processed/cleaned_reviews.parquet', index=False, compression = 'zstd')
```
