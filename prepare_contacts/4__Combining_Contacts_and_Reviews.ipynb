# Combining Contacs and Reviews

```python
import pandas as pd
```

```python
df_contacts = pd.read_parquet('../data/processed/Geocoded_Contacts.parquet')
df_contacts
```

```python
df_reviews = pd.read_parquet('../data/processed/cleaned_reviews.parquet',
                             columns = ['org_PAC_ID', 'patient_count','star_value'])
df_reviews = df_reviews.sort_values('org_PAC_ID', ignore_index=True)
df_reviews
```

```python
df_combined = pd.merge(df_contacts, df_reviews, how='inner', left_on='org_pac_id', right_on='org_PAC_ID')
df_combined = df_combined.reset_index(drop=True)
df_combined
```

```python
df_combined.to_parquet('../data/processed/Combined_Contacts_and_Reviews.parquet', index=False, compression = 'zstd')
```
