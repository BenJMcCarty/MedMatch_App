# Contact Cleaning and Subsetting

---

***Subsetting provider data for specific locations and for specific primary specializations.***

---

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/data.csv', dtype={'sec_spec_1': str, 'sec_spec_2': str, 'sec_spec_3': str, 'sec_spec_4': str, 'ZIP Code':str})
df = df.rename(columns=lambda x: x.strip())
df
```

```python
df = df.drop_duplicates(ignore_index=True)
df = df.dropna(subset = ['pri_spec', 'adr_ln_1'], how = 'any', ignore_index=True)
df
```

```python
df = df[df['State'] == 'MD'].reset_index(drop=True)
df
```

```python
pri_spec = list(df['pri_spec'].unique())
pri_spec.sort()
pri_spec
```

```python
selected_primary_specs = ['CHIROPRACTIC',
                          'EMERGENCY MEDICINE',
                          'FAMILY PRACTICE',
                          'GENERAL PRACTICE',
                          'NEUROLOGY',
                          'MENTAL HEALTH COUNSELOR',
                          'PAIN MANAGEMENT',
                          'PHYSICAL MEDICINE AND REHABILITATION'
 ]
df = df[df['pri_spec'].isin(selected_primary_specs)].reset_index(drop=True)
df
```

```python
df['ZIP Code'] = df['ZIP Code'].map(lambda x: x[:5])
df['ZIP Code']
```

```python
df['Full Address'] = df['adr_ln_1'] + ', ' + df['City/Town'] + ', ' + df['State'] + ' ' + df['ZIP Code']
df['Full Address']
```

```python
df.to_parquet('../data/processed/cleaned_contacts.parquet', index=False, compression = 'zstd')
```
